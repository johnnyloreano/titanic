```python
def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```

```python
def survived_bar(feature):
    survived = dataset[dataset['Survived'] == 1][feature].value_counts()
    dead = dataset[dataset['Survived'] == 0][feature].value_counts()
    return pd.DataFrame([dead,survived], index = ["Dead", "Survived"])
hide_toggle()
```

## Sex attribute

Can be [male,female]

```python
dataset['Sex'].value_counts().plot.pie( autopct='%1.0f%%', figsize = (5,5))
survived_bar('Sex').plot( kind = "pie", subplots=True, figsize = (10,10), autopct='%1.0f%%')
```

## Age attribute

```python
sns.violinplot(x='Sex', y='Age', 
               hue='Survived', data=dataset, 
               split=True,
               palette={0: "r", 1: "g"}
              );
```

## Pclass - Ticket Class 

Can be [1,2,3]. 1 is the most expensive, 3 is the most cheaper

```python
dataset['Pclass'].value_counts().plot.pie( autopct='%1.0f%%', figsize = (5,5))
survived_bar('Pclass').transpose().plot( kind = "pie", subplots=True, figsize = (10,10), autopct='%1.0f%%')
```

## Fare attribute

How much each one paid for the ticket

```python
sns.boxplot(x = 'Pclass', y = 'Fare', data = dataset, showfliers = True)
```

```python
from matplotlib.cbook import boxplot_stats  
outl = boxplot_stats(dataset.Fare).pop(0)['fliers']

outlier_data = dataset[(dataset['Fare'].isin(outl)) & (dataset['Pclass'] == 1)]

normal_data = dataset[(dataset['Fare'].isin(outl) == False) & (dataset['Pclass'] == 1)]

(outlier_data['Survived'].value_counts()).plot(kind = 'pie', autopct='%1.0f%%')
```
